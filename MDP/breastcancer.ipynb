```python
from runtest import *
```

```python
env = Data(unknown_rate=1)
env.loadfile("breastcancer.csv")
env.normalize()
env.alpha = 0
env.cluster_K_means(7)
n = len(env.data[0][1])

costs = None
if costs is None:
    env.set_costs()
else:
    env.set_costs(costs)
test_env = env.split(0.80)

res = []
for i in range(n):
    env.max_cost = (i+1)/n
    test_env.max_cost = env.max_cost
    res.append( runtest(env,test_env,max_eps=1000, epsilon_decay=0.999) )
    print()
    print("Q:",res[-1][0])
    print("R:",res[-1][1])
    print("N:",res[-1][2])
    print("M:",res[-1][3])
    
```

```python
Qs = [res[i][0] for i in range(len(res))]
Rs = [res[i][1] for i in range(len(res))]
Ns = [res[i][2] for i in range(len(res))]
Ms = [res[i][3] for i in range(len(res))]
print("Qs:",Qs)
print("Rs:",Rs)
print("Ns =",Ns)
print("Ms =",Ms)


Ts = [459.821, 343.446, 297.802, 290.812, 271.334, 255.672, 246.79, 244.666, 236.995]
Ns = [464.98311397996093, 422.99521738592074, 383.1262045745856, 317.6254178320944, 296.36685004439164, 297.983368066636, 264.93630664408147, 253.47948657279488, 205.62514052800202]
Ms = [508.2973101859528, 392.9049226985426, 345.4315361473615, 320.8759997814059, 290.17817386064655, 280.949065149959, 253.55999212075866, 225.1737985265104, 200.40819482505995]

import matplotlib.pyplot as plt
x = [(i+1)/n for i in range(n)]
plt.plot(x, Ns, x, Ms,x,Ts)
plt.show()
```
