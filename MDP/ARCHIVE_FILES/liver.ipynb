```python
env = Data(unknown_rate=1)
env.loadfile("liver.csv")
env.normalize()
env.alpha = 0
env.cluster_K_means(7)

costs = None
if costs is None:
    env.set_costs()
else:
    env.set_costs(costs)
test_env = env.split(0.80)

res = []
for i in range(10):
    env.max_cost = (i+1)/10
    test_env.max_cost = env.max_cost
    res.append( runtest(env,test_env,max_eps=1000, epsilon_decay=0.999) )
    print()
    print("Q:",res[-1][0])
    print("R:",res[-1][1])
    print("N:",res[-1][2])
    print("M:",res[-1][3])
    
```
