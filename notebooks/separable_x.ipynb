```python
import numpy as np
import separable_x as dg
import matplotlib.pylab as plt
import random
from clusteringtree import ClusteringTree

random.seed()
```

```python
points, labels = dg.gen(n=22) # generates 8 clusters
m1 = max([ points[i][0] for i in range(len(points)) ])
m2 = max([ points[i][1] for i in range(len(points)) ])
for i in range(len(points)):
    points[i][0] /= m1
    points[i][1] /= m2
x_p = [ points[i][0] for i in range(len(points)) ]
y_p = [ points[i][1] for i in range(len(points)) ]
```

```python
plt.plot(x_p,y_p,'o')
```

```python
data1 = np.array(points)
plt.plot(data1[:,0],data1[:,1],'o')
CT = ClusteringTree(np.linspace(0,1))
CT.cluster_multivariables(data1)
```

```python

plt.plot(data1[:,0],data1[:,1],'o')
CT.plot_boundries2D(y_upper_lim = 1., y_lower_lim = 0.8, x_left_lim = 0., x_right_lim = 1.)
plt.show()
CT.boundary
```
