```python
import numpy as np
import data_gen1 as dg
import matplotlib.pylab as plt
import random
from clusteringtree import ClusteringTree

random.seed(22)
```

```python
points, labels = dg.gen(n=8) # generates 8 clusters
x_p = [ points[i][0] for i in range(len(points)) ]
y_p = [ points[i][1] for i in range(len(points)) ]
x_id = np.linspace(0,20)
y_id = np.linspace(0,20)
```

```python
plt.plot(x_p,y_p,'o',x_id,y_id,'-')
plt.show()
```

First, let's construct the visualize the clusters with different colors.

```python
# find points with label = 1
data1 = []
for i in range(len(points)):
    if labels[i] == 1:
        data1.append(points[i])
data1=np.array(data1)

# find points with label = 0
data0=[]
for i in range(len(points)):
    if labels[i] == 0:
        data0.append(points[i])
data0=np.array(data0)
plt.plot(data1[:,0],data1[:,1],'o')
plt.plot(data0[:,0],data0[:,1],'o')
plt.show()
```

Superimposing the algorithm on clusters with label 1:

```python
plt.plot(data1[:,0],data1[:,1],'o')
plt.plot(data0[:,0],data0[:,1],'o')

CT = ClusteringTree()
CT.cluster_multivariables(data1)
CT.plot_boundries2D(y_upper_lim = 25., y_lower_lim = 0, x_left_lim = 0., x_right_lim = 25.)
plt.show()
```

And likewise, on data with label 0:

```python
plt.plot(data1[:,0],data1[:,1],'o')
plt.plot(data0[:,0],data0[:,1],'o')

CT = ClusteringTree()
CT.cluster_multivariables(data0)
CT.plot_boundries2D(y_upper_lim = 25., y_lower_lim = 0, x_left_lim = 0., x_right_lim = 25.)
plt.show()
```
