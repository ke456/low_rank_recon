```python
import numpy as np
import data_5_param as dg
import matplotlib.pylab as plt
import random
from clusteringtree_centered import ClusteringTree_centered
from mpl_toolkits import mplot3d
%matplotlib notebook

random.seed()

def avg(d):
    # Returns the average of d=List[Integer], 0 if the list is empty
    if (len(d)==0):
        return 0
    return [sum(d[:,i])/len(d) for i in range(len(d[0]))]
```

```python
points, labels = dg.gen() # generates 8 clusters
# normalize the data
m1 = max([ points[i][0] for i in range(len(points)) ])
m2 = max([ points[i][1] for i in range(len(points)) ])
m3 = max([ points[i][2] for i in range(len(points)) ])
for i in range(len(points)):
    points[i][0] /= m1
    points[i][1] /= m2
    points[i][2] /= m3
```

```python
x1_p = [ points[i][0] for i in range(len(points)) ]
x2_p = [ points[i][1] for i in range(len(points)) ]
x3_p = [ points[i][2] for i in range(len(points)) ]

d1 = []
d2 = []
for d in points:
    if d[0] < 0.95:
        d1.append(d)
    else:
        d2.append(d)
c1 = avg(np.array(d1))
c2 = avg(np.array(d2))

print("c1:",c1)
print("c2:",c2)

ax = plt.axes(projection='3d')

ax.set_xlabel('grade')
ax.set_ylabel('academic EC')
ax.set_zlabel('other EC')
ax.scatter3D(x1_p, x2_p, x3_p);

plt.show()
```

```python
#plt.plot(x1_p,x3_p,'o')
ax.scatter3D(c1[0],c1[1], c1[2]);
ax.scatter3D(c2[0],c2[1], c2[2]);
#plt.show()
```

```python
t = []
for d in np.array(d2):
    to_c1 = np.linalg.norm(d-c1)
    to_c2 = np.linalg.norm(d-c2)
    if to_c1 <= to_c2:
        t.append(d)
t = np.array(t)        
print("t:",t)
ax.scatter3D(t[:,0],t[:,1], t[:,2]);
```
