```python
import numpy as np
import data_5_param as dg
import matplotlib.pylab as plt
import random
from clusteringtree import ClusteringTree
from mpl_toolkits import mplot3d
%matplotlib notebook

random.seed()
```

```python
points, labels = dg.gen() # generates 8 clusters
# normalize the data
m1 = max([ points[i][0] for i in range(len(points)) ])
m2 = max([ points[i][1] for i in range(len(points)) ])
m3 = max([ points[i][2] for i in range(len(points)) ])
for i in range(len(points)):
    points[i][0] /= m1
    points[i][1] /= m2
    points[i][2] /= m3
```

```python
x1_p = [ points[i][0] for i in range(len(points)) ]
x2_p = [ points[i][1] for i in range(len(points)) ]
x3_p = [ points[i][2] for i in range(len(points)) ]

ax = plt.axes(projection='3d')

ax.set_xlabel('grade')
ax.set_ylabel('academic EC')
ax.set_zlabel('other EC')
ax.scatter3D(x1_p, x2_p, x3_p);
```

```python
plt.plot(x1_p,x3_p,'o')
plt.show()
```
