```python
import matplotlib.pyplot as plt

def read(fname):
    f = open(fname,"r")
    line = f.readline()
    s = line.split(',')
    data = []
    for ss in s:
        data.append(float(ss))
    return data

budgets = [(i+1)/10 for i in range(10)]
```

```python
# for hcv
test = "hcv"

tree = read("csv_files/"+test+"_tree.csv")
rl = read("csv_files/"+test+"_RL.csv")
rand = read("csv_files/"+test+"_rand.csv")

plt.plot(budgets, tree, 'r', budgets, rl, 'b', budgets, rand, 'g')
plt.show()
```
